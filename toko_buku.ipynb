```python
import json
import os

# File penyimpanan
FILE_BUKU = "data_buku.json"
FILE_TRANSAKSI = "transaksi.json"

# Cek dan load data buku
if os.path.exists(FILE_BUKU):
    with open(FILE_BUKU, "r") as f:
        daftar_buku = json.load(f)
else:
    daftar_buku = []

# Cek dan load data transaksi
if os.path.exists(FILE_TRANSAKSI):
    with open(FILE_TRANSAKSI, "r") as f:
        transaksi = json.load(f)
else:
    transaksi = []

# Fungsi simpan ke file
def simpan_data():
    with open(FILE_BUKU, "w") as f:
        json.dump(daftar_buku, f, indent=4)
    with open(FILE_TRANSAKSI, "w") as f:
        json.dump(transaksi, f, indent=4)

# Tambah buku baru
def tambah_buku():
    judul = input("Judul buku: ")
    penulis = input("Penulis: ")
    try:
        harga = int(input("Harga: "))
        stok = int(input("Stok: "))
        buku = {
            "judul": judul,
            "penulis": penulis,
            "harga": harga,
            "stok": stok
        }
        daftar_buku.append(buku)
        simpan_data()
        print(f"Buku '{judul}' berhasil ditambahkan!\n")
    except ValueError:
        print("Harga dan stok harus angka!\n")

# Lihat daftar buku
def lihat_buku():
    if not daftar_buku:
        print("Belum ada buku yang tersedia.\n")
        return
    print("\nDaftar Buku Toko:")
    for i, buku in enumerate(daftar_buku, 1):
        print(f"{i}. {buku['judul']} - {buku['penulis']} | Harga: {buku['harga']} | Stok: {buku['stok']}")
    print()

# Jual buku
def jual_buku():
    lihat_buku()
    if not daftar_buku:
        return
    try:
        indeks = int(input("Pilih nomor buku yang dijual: ")) - 1
        if indeks < 0 or indeks >= len(daftar_buku):
            print("Nomor tidak valid.\n")
            return
        jumlah = int(input("Jumlah yang dijual: "))
        buku = daftar_buku[indeks]
        if jumlah > buku['stok']:
            print("Stok tidak cukup!\n")
            return
        buku['stok'] -= jumlah
        transaksi.append({
            "judul": buku['judul'],
            "jumlah": jumlah,
            "total": buku['harga'] * jumlah
        })
        simpan_data()
        print(f"Berhasil menjual {jumlah} buku '{buku['judul']}'!\n")
    except ValueError:
        print("Input harus angka!\n")

# Menu utama
def menu():
    while True:
        print("=== APLIKASI TOKO BUKU ===")
        print("1. Tambah Buku")
        print("2. Lihat Daftar Buku")
        print("3. Transaksi Penjualan")
        print("4. Keluar")
        pilih = input("Pilih menu (1-4): ")

        if pilih == "1":
            tambah_buku()
        elif pilih == "2":
            lihat_buku()
        elif pilih == "3":
            jual_buku()
        elif pilih == "4":
            print("Terima kasih telah menggunakan Aplikasi Toko Buku!")
            break
        else:
            print("Pilihan tidak valid!\n")

menu()
```
